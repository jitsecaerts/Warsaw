## Exploratory Data Analysis (EDA): Credit Card Fraud

### Introduction

In this notebook, we perform an extensive Exploratory Data Analysis (EDA) on the Credit Card Fraud Detection dataset.The goal is to:
- Understand the structure of the PCA-transformed features.
- Assess data quality (detecting outliers and skewness in 'Time' and 'Amount').Analyze the extreme class imbalance (Fraud vs. Normal).
- Identify critical correlations between latent features ($V1-V28$) and the target class.

### Importing Libraries and Loading the Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')

data = pd.read_csv("creditcard.csv")
data.head()
```

### Target Variable Analysis (Class)

#### Class Distribution

In fraud detection, accuracy is misleading because the target is highly imbalanced.

```python
plt.figure(figsize=(7,5))
sns.countplot(x='Class', data=data)
plt.title('Fraud Distribution (0: Normal, 1: Fraud)')
plt.yscale('log') 
plt.show()

print(f"Percentage of Fraud: {round(data['Class'].value_counts()[1]/len(data)*100, 3)}%")
```

### Univariate Analysis (Distributions)

#### Numerical Feature Distributions (Time & Amount)

We focus on the non-PCA features to see if they require scaling.

```python
fig, ax = plt.subplots(1, 2, figsize=(18,4))

sns.histplot(data['Amount'], ax=ax[0], color='r', kde=True)
ax[0].set_title('Distribution of Transaction Amount')

sns.histplot(data['Time'], ax=ax[1], color='b', kde=True)
ax[1].set_title('Distribution of Transaction Time')

plt.show()
```

### PCA Features `($V1-V28$)`

We visualize a subset of PCA features to check for normality.

```python
v_features = data.iloc[:, 1:29].columns
plt.figure(figsize=(12, 8))
for i, col in enumerate(v_features[:4]):
    plt.subplot(2, 2, i+1)
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

### Feature vs Target Analysis

#### Boxplots: Feature distributions by Class

This helps identify which PCA components are the strongest "signals" for fraud.

```python
cols = ['V17', 'V14', 'V12', 'V10']
for col in cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x='Class', y=col, data=data)
    plt.title(f'{col} vs Class (Fraud Signal)')
    plt.show()
```

### Correlation Matrix

```python
plt.figure(figsize=(12,8))
corr = data.corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix (PCA Components)')
plt.show()
```

### Most Correlated Features with Fraud

```python
corr_target = corr['Class'].sort_values(ascending=False)
plt.figure(figsize=(10,6))
corr_target.drop('Class').plot(kind='bar', color='darkred')
plt.title('Feature Correlation with Fraud (Class)')
plt.ylabel('Correlation Coefficient')
plt.show()
```

### Advanced Distribution Analysis (Skewness)

In credit datasets, the 'Amount' feature is usually heavily skewed. Most transactions are small, but a few are very large. This skewness can negatively impact the convergence of Neural Networks.

```python
from scipy.stats import skew

print(f"Skewness of Amount: {skew(data['Amount'])}")
print(f"Skewness of Time: {skew(data['Time'])}")

plt.figure(figsize=(8,6))
sns.kdeplot(data['Amount'], color="red", fill=True)
plt.title('KDE of Amount (Showing Heavy Right Skew)')
plt.show()
```
